```python
# Derived feature: price_difference
df['price_difference'] = df['actual_price'] - df['discounted_price']

# Normalize rating scores (Min-Max normalization)
df['normalized_rating'] = (df['rating'] - df['rating'].min()) / (df['rating'].max() - df['rating'].min())

# Derived feature: value_for_money_score (normalized rating divided by price_difference)
df['value_for_money_score'] = df['normalized_rating'] / (df['price_difference'] + 1)  # +1 to avoid division by zero

# Derived feature: weighted_rating (rating * log1p(rating_count))
df['weighted_rating'] = df['rating'] * np.log1p(df['rating_count'])


df['weighted_rating'] = (df['weighted_rating'] / df['weighted_rating'].max()) * 5
```

```python
def fix_price_outliers(df, price_col, category_col, new_col='actual_price_fixed'):
    def process_group(group):
        median = group[price_col].median()
        mad = np.median(np.abs(group[price_col] - median))
        
        if mad == 0:
            z = pd.Series(0, index=group.index)  # make it a Series
        else:
            z = 0.6745 * (group[price_col] - median) / mad
        
        threshold = 3
        max_valid = group.loc[np.abs(z) <= threshold, price_col].max()
        fixed_prices = group[price_col].copy()
        fixed_prices[np.abs(z) > threshold] = max_valid
        group[new_col] = fixed_prices
        return group

    return df.groupby(category_col, group_keys=False).apply(process_group)

df = fix_price_outliers(df, 'actual_price', 'category')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

top_categories = df['category'].value_counts().nlargest(10).index
fig, axes = plt.subplots(5, 2, figsize=(30, 60), sharex=False, sharey=False)
axes = axes.flatten()

for i, cat in enumerate(top_categories):
    sns.histplot(df[df['category'] == cat]['actual_price_fixed'], bins=30, ax=axes[i], kde=True)
    axes[i].set_title(cat, fontsize=16)  # Increase font size as needed
    axes[i].set_xlabel('Actual Price Fixed Outliers')
    axes[i].set_ylabel('Frequency')

# plt.tight_layout()
plt.show()
```
