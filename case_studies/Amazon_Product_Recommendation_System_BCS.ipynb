```python
df['discounted_price'] = df['discounted_price'].str.replace('₹', '').str.replace(',', '')
df['actual_price'] = df['actual_price'].str.replace('₹', '').str.replace(',', '')
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '')
df['rating_count'] = df['rating_count'].str.replace(',', '')
```

```python
df['discounted_price'] = df['discounted_price'].astype(float)
df['actual_price'] = df['actual_price'].astype(float)
df['discount_percentage'] = df['discount_percentage'].astype(float)
```

```python
# df['rating'] = df['rating'].astype(float)

# ValueError: could not convert string to float: '|'

df[df['rating'] == '|']
```

```python
df["rating"] = df['rating'].apply(lambda x:  np.nan if x == '|' else float(x))
```

```python
df[df['rating'] == '|']
```

```python
# df['rating_count'] = df['rating_count'].astype(int)    

# ValueError: cannot convert float NaN to integer happens because standard NumPy/Pandas int64 dtype cannot represent NaN.
# 👉 Solution is to use Pandas' nullable integer dtype (Int64), which supports NaN (stored as <NA>).

df['rating_count'] = df['rating_count'].astype("Int64")    
```

```python
# Remove outliers using the IQR method
Q1 = df['actual_price'].quantile(0.25)
Q3 = df['actual_price'].quantile(0.75)
IQR = Q3 - Q1

filtered_df = df[(df['actual_price'] >= Q1 - 1.5 * IQR) & (df['actual_price'] <= Q3 + 1.5 * IQR)]['actual_price']
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(filtered_df, bins=200, kde=True)
plt.title('Distribution of Actual Prices')
plt.xticks(rotation=80)
# plt.locator_params(axis='x', nbins=50) tells Matplotlib to try to place up to 50 ticks on the x-axis.
plt.locator_params(axis='x', nbins=40)
plt.xlabel('Actual Price')
plt.ylabel('Frequency')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Remove outliers using the IQR method
Q1 = df['actual_price'].quantile(0.25)
Q3 = df['actual_price'].quantile(0.75)
IQR = Q3 - Q1
filtered_df = df[(df['actual_price'] >= Q1 - 1.5 * IQR) & (df['actual_price'] <= Q3 + 1.5 * IQR)]['actual_price']

plt.figure(figsize=(12, 6))
sns.histplot(filtered_df, bins='auto', kde=True)
plt.title('Distribution of Actual Prices (Outliers Removed)')
plt.xlabel('Actual Price')
plt.ylabel('Frequency')
plt.xticks(rotation=80)
plt.locator_params(axis='x', nbins=40)
plt.grid(True, linestyle='--', alpha=0.5)
plt.axvline(filtered_df.mean(), color='red', linestyle='dashed', linewidth=1, label='Mean')
plt.axvline(filtered_df.median(), color='green', linestyle='dashed', linewidth=1, label='Median')
plt.legend()
plt.tight_layout()
plt.show()

# Log scale version
plt.figure(figsize=(12, 6))
sns.histplot(filtered_df, bins='auto', kde=True)
plt.xscale('log')
plt.title('Distribution of Actual Prices (Log Scale, Outliers Removed)')
plt.xlabel('Actual Price (Log Scale)')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
